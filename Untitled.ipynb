<h1>4. Data splitting</h1>
<p>We are going to split the data into training and testing sets. I will do 70 to 30 split.</p>

```python
from sklearn.model_selection import train_test_split

data = data.drop(columns="Index")
X = data.drop(columns="class")
y = data["class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

<h1>5. Model training</h1>
<p>Supervised learning is a type of machine learning where an algorithm is trained to make predictions based on labeled data. In supervised learning, we use labeled data to teach the computer how to make predictions about new, unseen data. The algorithm learns from the labeled data by identifying patterns and relationships between the input features and the labeled outcomes. Once the algorithm has learned these patterns, it can make predictions about new data by applying these patterns to the new input features.</p>

There are two types of supervised learning - classification and regression.

Classification is a type of supervised learning where the goal is to predict which category or class an input example belongs to. In other words, we are trying to assign a label or category to each input example based on its features or attributes.

For example, in email spam detection, we might use classification to predict whether a new email is spam or not spam based on its subject, sender, and content. In medical diagnosis, we might use classification to predict whether a patient has a particular disease based on their symptoms and medical history.

Classification problems can be either binary or multi-class. Binary classification involves predicting one of two possible outcomes, such as "spam" or "not spam", while multi-class classification involves predicting one of more than two possible outcomes, such as "cat", "dog", or "bird".

Our data falls into the category of <b>binary classification</b>

<h1>Defining a helper function</h1>
<p>Evaluation of classification models is typically done using metrics such as accuracy, precision, recall, and F1 score. These metrics help to assess the performance of the algorithm and identify areas where it may need improvement.
</P>
<p>We will create a helper function where all of our result from different algorithms are going to be stored for future comparison.</p>

```python
model_name = []
accuracy_result  = []
f1_result = []
recall_result = []
precision_result = []

def algo_holder(model, acc, f1, recall, precision):
    if model not in model_name:
        model_name.append(model)
    accuracy_result.append(round(acc,5))
    f1_result.append(round(f1,5))
    recall_result.append(round(recall,5))
    precision_result.append(round(precision,5))
```
